   - Create a list of numbers from 1 to 20
    numbers = list(range(1, 21))


### Exercise 5 – Functions
 Write a function that receives a list of numbers
 and returns the average value
```python
def calculate_average(numbers):
    pass
```
Test it with:
scores = [12, 15, 18, 10, 14]



````python
# exercice 5
def avg(numbers_list):
    som = 0
    for nbr in numbers_list: som += float(nbr)
    return som /len(numbers_list)

scores = [12, 15, 18, 10, 14]
print(avg(scores))
````


### Exercise 6 – Combining Everything
 Given a list of students' scores
 scores = [45, 67, 89, 34, 90, 56, 78]

1. Keep only scores >= 60
2. Calculate the average of the successful scores
3. Print the result



Use:

loops OR list comprehension

your function from Exercise 5

````python
# exercice 6
scores = [45, 78, 90, 62, 55, 88, 39]

successful_scores = [score for score in scores if score >= 60]

print("ssuccessful scores: ", successful_scores)

# average
avg_score = avg(successful_scores)

for score in successful_scores:
    print(score if score > avg_score else "", end=" ")
    
````
